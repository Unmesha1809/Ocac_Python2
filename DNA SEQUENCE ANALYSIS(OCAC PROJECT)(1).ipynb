```python
import matplotlib.pyplot as plt
from Bio import Entrez, SeqIO
from collections import Counter
from urllib.error import HTTPError
import pandas as pd

# Set your email address for NCBI
Entrez.email = "your_email@example.com"

def fetch_dna_sequences(accession_numbers):
    records = []
    for accession in accession_numbers:
        try:
            # Fetch the DNA sequence from NCBI
            handle = Entrez.efetch(db="nucleotide", id=accession, rettype="gb", retmode="text")
            record = SeqIO.read(handle, "genbank")
            handle.close()
            records.append(record)
        except HTTPError as e:
            print(f"HTTP Error for {accession}: {e}")
    return records

def count_nucleotides(sequence):
    nucleotide_counts = Counter(sequence)
    return nucleotide_counts

def create_nucleotide_dataframe(nucleotide_counts_list):
    # Create a list of dataframes for each sequence
    df_list = []
    for idx, counts in enumerate(nucleotide_counts_list):
        df = pd.DataFrame({
            'Nucleotide': ['A', 'T', 'G', 'C'],
            f'Sequence_{idx+1}': [counts.get(n, 0) for n in ['A', 'T', 'G', 'C']]
        })
        df_list.append(df)
    
    # Merge all dataframes into a single dataframe
    df_final = df_list[0]
    for df in df_list[1:]:
        df_final = df_final.merge(df, on='Nucleotide')
    
    return df_final

def plot_nucleotide_distribution(df):
    # Plot the nucleotide distribution
    df.set_index('Nucleotide').plot(kind='bar')
    plt.xlabel('Nucleotides')
    plt.ylabel('Frequency')
    plt.title('Nucleotide Distribution in DNA Sequences')
    plt.show()

# Example usage
accession_numbers = ["NM_001301717", "NM_001301718"]  # Replace with your actual accession numbers
dna_sequence_records = fetch_dna_sequences(accession_numbers)

if dna_sequence_records:
    nucleotide_counts_list = [count_nucleotides(record.seq) for record in dna_sequence_records]
    nucleotide_df = create_nucleotide_dataframe(nucleotide_counts_list)
    
    print(nucleotide_df)
    plot_nucleotide_distribution(nucleotide_df)
else:
    print("Failed to fetch the DNA sequences.")
```
